```python
import pandas as pd
```

```python
dataset = pd.read_csv(
        filepath_or_buffer = "../dataset/Dataset.csv",
        header = 11,
        skiprows = [12],
        index_col = False,
        low_memory = False)
```

```python
dataset = dataset[["ID", "SecType", "Date", "Time", "Last", "Trading time", "Trading date"]]
dataset = dataset.rename(columns = {"Trading date" : "TradingDate", "Trading time" : "TradingTime"})
dataset
```

```python
nullDate = dataset[dataset["Date"].isnull()]["Date"].count()
nullTime = dataset[dataset["Time"].isnull()]["Time"].count()
nullTradingTime = dataset[dataset["TradingTime"].isnull()]["TradingTime"].count()
nullTradingDate = dataset[dataset["TradingDate"].isnull()]["TradingDate"].count()

(nullDate, nullTime, nullTradingDate, nullTradingTime)
```

```python
cleanedDataset = dataset
cleanedDataset = dataset[dataset["TradingTime"] != "00:00:00.000"]
cleanedDataset = cleanedDataset[cleanedDataset["TradingDate"].notnull()]
cleanedDataset
```

```python
frNum = cleanedDataset[cleanedDataset["ID"].str.endswith(".FR")]["ID"].count()
nlNum = cleanedDataset[cleanedDataset["ID"].str.endswith(".NL")]["ID"].count()
etrNum = cleanedDataset[cleanedDataset["ID"].str.endswith(".ETR")]["ID"].count()

(frNum, nlNum, etrNum)
```

```python
nullLastNum = cleanedDataset[cleanedDataset["Last"].isnull()]["Last"].count()
nullLastNum
```
